  (a) Write a program (in any programming language) to simulate the tossing of a 10-sided fair die (i.e., all sides are equally likely),
 for t = 50, 100, 1000, 2000, 3000, 10000, and 100000 tosses.
Based on the simulation, what is the estimated probability of obtaining an odd number?
  
  (b) Suppose X is a random variable denoting the outcome of a die toss. Based on mathematical analysis and probability theory, what is the probability that X has an odd value?
  
  (c) Refer back to part (a). Does it agree with the theoretical result in (b)?

  (d) Repeat parts (a), (b), and (c) with a 10-sided die that has the property that any odd number is twice as likely as an even number and all even numbers are equally likely.

(a)

```python
import random

t_values = [50, 100, 1000, 2000, 3000, 10000, 100000]

for t in t_values:
    odd_count = 0
    for i in range(t):
        toss_result = random.randint(1, 10)
        if toss_result % 2 == 1:
            odd_count += 1
    odd_probability = odd_count / t
    print(f"For {t} tosses, the probability of obtaining an odd number is: {odd_probability}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Define the interval for the uniform distribution
a = 3
b = 7

# Define the values of n to consider
N = [1, 2, 3, 10, 30, 100]

# Define the number of samples
t = 10**4

# Generate t samples of Zn for each value of n
samples = {}
for n in N:
    # Generate t samples of Xi for each value of n
    X = uniform.rvs(loc=a, scale=b-a, size=(n, t))

    # Compute Zn for each sample of Xi
    Zn = np.mean(X, axis=0)

    # Store the samples of Zn
    samples[n] = Zn

# Plot the histograms of the samples of Zn
fig, axs = plt.subplots(2, 3, figsize=(12, 8))
axs = axs.flatten()
for i, n in enumerate(N):
    axs[i].hist(samples[n], density=True, bins=50)
    axs[i].set_xlabel('Zn')
    axs[i].set_ylabel('PDF')
    axs[i].set_title('PDF of Zn for n = {}'.format(n))

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform, norm

# Define the interval for the uniform distribution
a = 3
b = 7

# Define the values of n to consider
N = [1, 2, 3, 10, 30, 100]

# Define the number of samples
t = 10**4

# Generate t samples of Zn and Y for each value of n
samples_Zn = {}
samples_Y = {}
for n in N:
    # Generate t samples of Xi for each value of n
    X = uniform.rvs(loc=a, scale=b-a, size=(n, t))

    # Compute Zn for each sample of Xi
    Zn = np.mean(X, axis=0)

    # Compute Y with the same mean and variance as Zn
    mu_Y = np.mean(X)
    var_Y = np.var(X) / n
    std_Y = np.sqrt(var_Y)
    Y = norm.rvs(loc=mu_Y, scale=std_Y, size=t)

    # Store the samples of Zn and Y
    samples_Zn[n] = Zn
    samples_Y[n] = Y

# Plot the histograms of the samples of Zn and Y with their respective PDFs
fig, axs = plt.subplots(2, 3, figsize=(12, 8))
axs = axs.flatten()
for i, n in enumerate(N):
    # Plot the histogram of Zn
    axs[i].hist(samples_Zn[n], density=True, bins=50, alpha=0.5)

    # Plot the PDF of Zn
    x = np.linspace(np.min(samples_Zn[n]), np.max(samples_Zn[n]), 100)
    pdf_Zn = norm.pdf(x, loc=np.mean(samples_Zn[n]), scale=np.std(samples_Zn[n]))
    axs[i].plot(x, pdf_Zn, 'r', linewidth=2, label='PDF of Zn')

    # Plot the histogram of Y
    axs[i].hist(samples_Y[n], density=True, bins=50, alpha=0.5)

    # Plot the PDF of Y
    pdf_Y = norm.pdf(x, loc=np.mean(samples_Y[n]), scale=np.std(samples_Y[n]))
    axs[i].plot(x, pdf_Y, 'g', linewidth=2, label='PDF of Y')

    axs[i].set_xlabel('Zn and Y')
    axs[i].set_ylabel('PDF')
    axs[i].set_title('PDF of Zn and Y for n = {}'.format(n))
    axs[i].legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
a = 3
b = 7
ns = [1, 2, 3, 10, 30, 100]  # different values of n

# Generate t samples of Zn for each value of n
t = 10000  # number of samples
for n in ns:
    # Generate t samples of Xi uniformly distributed in (a,b)
    Xi = np.random.uniform(a, b, (t, n))

    # Compute Zn for each sample
    Zn = np.mean(Xi, axis=1)

    # Compute the mean and variance of Zn
    mu = (a+b)/2
    sigma2 = (b-a)**2/12

    # Generate x values for plotting the PDF
    x = np.linspace(mu-3*np.sqrt(sigma2), mu+3*np.sqrt(sigma2), 1000)

    # Compute the PDF values for each x value
    fx = np.exp(-(x-mu)**2/(2*sigma2))/np.sqrt(2*np.pi*sigma2)

    # Plot the histogram of Zn and the PDF of the normal distribution
    plt.hist(Zn, bins=30, density=True, alpha=0.5)
    plt.plot(x, fx, 'r', linewidth=2)
    plt.xlabel('Zn')
    plt.ylabel('PDF')
    plt.title('PDF of Zn for n={}'.format(n))
    plt.legend(['PDF', 'Histogram'])
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the interval for the uniform distribution
a = 3
b = 7

# Define the values of n to consider
N = [1, 2, 3, 10, 30, 100]

# Define the number of points to plot the exact PDF of Zn
num_points = 100

# Define the function to compute the exact PDF of Zn
def exact_pdf_Zn(x, n):
    mu = (a + b) / 2
    var = ((b - a)**2) / (12 * n)
    std = np.sqrt(var)
    return norm.pdf(x, loc=mu, scale=std)

# Plot the exact PDF of Zn for each value of n
fig, axs = plt.subplots(2, 3, figsize=(12, 8))
axs = axs.flatten()
for i, n in enumerate(N):
    # Define the range of Zn to plot
    x = np.linspace(a, b, num_points)

    # Compute the exact PDF of Zn
    pdf_Zn = exact_pdf_Zn(x, n)

    # Plot the exact PDF of Zn
    axs[i].plot(x, pdf_Zn, 'r', linewidth=2, label='Exact PDF of Zn')

    # Superimpose the histograms of Zn and Y with their respective PDFs
    # Generate t samples of Xi for each value of n
    t = 10**4
    X = np.random.uniform(low=a, high=b, size=(n, t))

    # Compute Zn for each sample of Xi
    Zn = np.mean(X, axis=0)

    # Compute Y with the same mean and variance as Zn
    mu_Y = np.mean(X)
    var_Y = np.var(X) / n
    std_Y = np.sqrt(var_Y)
    Y = np.random.normal(loc=mu_Y, scale=std_Y, size=t)

    # Plot the histograms of the samples of Zn and Y with their respective PDFs
    axs[i].hist(Zn, density=True, bins=50, alpha=0.5)
    axs[i].hist(Y, density=True, bins=50, alpha=0.5)

    # Compute and plot the PDF of Zn and Y
    pdf_Zn = norm.pdf(x, loc=np.mean(Zn), scale=np.std(Zn))
    axs[i].plot(x, pdf_Zn, 'g', linewidth=2, label='PDF of Zn')
    pdf_Y = norm.pdf(x, loc=np.mean(Y), scale=np.std(Y))
    axs[i].plot(x, pdf_Y, 'b', linewidth=2, label='PDF of Y')

    axs[i].set_xlabel('Zn and Y')
    axs[i].set_ylabel('PDF')
    axs[i].set_title('PDF of Zn and Y for n = {}'.format(n))
    axs[i].legend()

plt.tight_layout()
plt.show()
```
